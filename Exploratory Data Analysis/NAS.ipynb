```python
# Filter out the 11 years age group and drop rows witdf_h missing values in relevant columns
df_filtered = df[(df["Age"] != "11 years") & df["Father.edu"].notna() & df["Science.."].notna()]
```

```python
df_filtered.head()
```

```python
# Group by Father's Education and Age, then calculate average science marks
average_marks = df_filtered.groupby(["Father.edu", "Age"])["Science.."].mean().reset_index()
average_marks
```

```python
# Calculate the overall average science marks per age group
age_group_avg = df_filtered.groupby("Age")["Science.."].mean().reset_index()
age_group_avg.rename(columns={"Science..": "Age_Group_Avg"}, inplace=True)
age_group_avg
```

```python
# Merge the average marks with overall age group averages
merged_data = average_marks.merge(age_group_avg, on="Age")

# Add a column to indicate if the performance is significantly better than average
merged_data["Better_Than_Avg"] = merged_data["Science.."] > merged_data["Age_Group_Avg"]

print(merged_data)
```
