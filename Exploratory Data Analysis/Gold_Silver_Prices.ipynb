```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
df = pd.read_csv("EDA_Gold_Silver_prices.csv")
```

```python
df.head()
```

```python
plt.scatter(df["SilverPrice"], df["GoldPrice"])
plt.xlabel("Silver Price")
plt.ylabel("Gold Price")
plt.title("Silver Price vs Gold Price")
plt.show()
```

```python
correlation = df["SilverPrice"].corr(df["GoldPrice"])
print(f"Correlation between SilverPrice and GoldPrice: {correlation}")
```

```python
df.head()
```

```python
# Extract the year from the Month column
df["Year"] = df["Month"].str[-2:]  # Extract last two digits of the year
df["Year"] = "20" + df["Year"]  # Add '20' prefix for 2000s
df["Year"] = df["Year"].astype(int)  # Convert to integer for comparison

# Filter data for 2008
df_2008 = df[df["Year"] == 2008]
```

```python
df_2008.head()
```

```python
correlation_2008 = df_2008["SilverPrice"].corr(df_2008["GoldPrice"])
```

```python
correlation_2008
```

```python
# Compute correlation for 2008
correlation = df_2008["SilverPrice"].corr(df_2008["GoldPrice"])
print(f"Correlation between Gold and Silver prices in 2008: {correlation:.2f}")
```
