### Problem Statement:

##### Consider that a real estate company has the data of real estate prices in Delhi. 
##### The company wants to optimise the selling price of the properties, based on important factors such as area, bedrooms, parking, etc.

### Essentially, the company wants:

##### To identify the variables affecting house prices, e.g., area, number of rooms, bathrooms, etc.
##### To create a linear model that quantitatively relates house prices with variables, such as the number of rooms, area, number of bathrooms, etc.
##### To know the accuracy of the model, i.e. how well do these variables predict the house prices

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings('ignore')
```

## Step-2: Preparing the Data for Modelling
- Encoding
    - Converting binary var 1/0
    - Other categorical vars to dummy vars
- Splitting in Test and Train data
- Rescaling of variable

```python
# yes/no Varaible 

varlist = ['airconditioning', 'hotwaterheating', 'mainroad', 'guestroom', 'basement', 'prefarea']


housing[varlist] = housing[varlist].apply(lambda x: x.map({'yes': 1, 'no': 0}))
housing[varlist].head()
```

```python
housing.head()
```

### Dummy Variables

```python
# Creating dummy cars for furnishingstatus

status = pd.get_dummies(housing['furnishingstatus']).astype(int)
status.head()
```

### We can also do like

- `00` -- `furnished`
- `01` -- `semi-furnished`
- `10` -- `unfurnished`  

```python
# Creating dummy cars for furnishingstatus
# dropping redundant vars

status = pd.get_dummies(housing['furnishingstatus'], drop_first=True).astype(int)
status.head()
```

```python
# concat the dummy vars

housing = pd.concat([housing, status], axis=1)
housing.head()
```

```python
housing = housing.drop('furnishingstatus', axis=1)
housing.head()
```

### Splitting data into train and test data

```python
df_train, df_test = train_test_split(housing, train_size=0.7, random_state=100)

print(df_train.shape)
print(df_test.shape)
```

### Rescaling the variables

```python
# 1. Instantiate the object

scaler = MinMaxScaler()

# Create a list of numeric vars

num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

# 2. Fit the data

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

## Step-3: Training the Model

```python
# Heatmap

plt.figure(figsize=(20,12))
sns.heatmap(df_train.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
df_train.head()
```

```python
# X_train, y_train

y_train = df_train.pop('price')
X_train = df_train
```

```python
# Add a constant
X_train_sm = sm.add_constant(X_train['area'])

# Create a first model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.params
```

```python
# Add another variable bathrooms
X_train_sm = X_train[['area', 'bathrooms']]
X_train_sm = sm.add_constant(X_train_sm)

# Create model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# Add another variable bedrooms
X_train_sm = X_train[['area', 'bathrooms', 'bedrooms']]
X_train_sm = sm.add_constant(X_train_sm)

# Create model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.summary()
```

### Adding all the variables to the Model

```python
housing.columns
```

```python
# Building the model with all variable
X_train_sm = sm.add_constant(X_train)

# Create model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# Significance (p-value)
# VIF - Variance Inflation Factor
```

### VIF - Variance Inflation Factor

VIFi = 1/(1-R2) 

If VIF = 5 then,

1-R2 = 0.2
R2 = o.8

Which is really High

```python
# Creating a DF which has variables and respective VIF

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by= 'VIF', ascending=False)
vif
```

#### We Could have
- High p-value, High VIF
- High-Low
    - High p-value, Low VIF -- will remove this first
    - Low p-value, High VIF -- then after that will remove these
- Low p-value, Low VIF

```python
X = X_train.drop('semi-furnished', axis=1)
```

```python
# Create another model
X_train_sm = sm.add_constant(X)

# Create model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# Creating a DF which has variables and respective VIF

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by= 'VIF', ascending=False)
vif
```

```python
X = X_train.drop('bedrooms', axis=1)
```

```python
# Create another model
X_train_sm = sm.add_constant(X)

# Create model
lr = sm.OLS(y_train,X_train_sm)

# Fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# Creating a DF which has variables and respective VIF

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by= 'VIF', ascending=False)
vif
```

### Step-4: Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm)
y_train_pred
```

```python
res = y_train - y_train_pred
sns.distplot(res)
```
